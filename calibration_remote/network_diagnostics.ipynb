```python
# %% [markdown]
# # Network Bottleneck Analysis
# 
# This notebook identifies bottlenecks in the occupational mobility network by analyzing:
# - Distribution of occupation sizes (target_demand)
# - Network connectivity patterns
# - Clusters of small occupations
# - Risk of zero vacancy creation
# - Expected flows (vacancies vs separations)

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from matplotlib.patches import Rectangle
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import squareform

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# %%
# Set up paths
path = "~/Documents/Documents - Nuff-Malham/GitHub/transition_abm/calibration_remote/"

nw = "full_omn"
#nw = "omn_soc_minor"
if nw == "full_omn":
    # Load all the data
    print("Loading data...")
    A = pd.read_csv("/Users/ebbamark/OneDrive - Nexus365/GenerateOccMobNets/data/asec_11_19_occ_alltransitions_2025_normalised.csv", header=None)
    ipums_input = pd.read_csv(path + "dRC_Replication/data/ipums_variables_full_omn_w_exp.csv", delimiter=",")
    occ_ids = pd.read_csv("/Users/ebbamark/OneDrive - Nexus365/GenerateOccMobNets/data/occ_names_employment_asec_occ.csv", delimiter=",")[['Code', 'Label']]
    occ_ids = occ_ids.rename(columns={"Label": "label"})
    occ_ids['id'] = occ_ids.index.values

    employment = np.round(ipums_input[['emp']]/10000) + 1
    unemployment = round(employment * (0.05 / 0.95))
    vac_rate_base = pd.read_csv(path + "dRC_Replication/data/vacancy_rateDec2000.csv").iloc[:, 2].mean() / 100
    vacancies = round(employment * vac_rate_base / (1 - vac_rate_base))
    demand_target = employment + vacancies

    print("Data loaded successfully!")

    # %%
    # Combine into analysis dataframe
    analysis_df = pd.DataFrame({
        'occ_id': range(len(A)),
        'label': occ_ids['label'].values,
        'code': occ_ids['Code'].values,
        'employment': employment.values.flatten(),
        'unemployment': unemployment.values.flatten(),
        'vacancies': vacancies.values.flatten(),
        'target_demand': demand_target.values.flatten(),
        'wage': ipums_input['median_weekly_earnings'].values * 52
    })

    print(f"Created analysis dataframe with {len(analysis_df)} occupations")

elif nw == "omn_soc_minor":

    # Load all the data
    print("Loading data...")
    A = pd.read_csv("/Users/ebbamark/OneDrive - Nexus365/GenerateOccMobNets/data/omn_asec_11_19_soc_2010_minor_alltransitions_2025_normalised.csv", header = None)
    ipums_input = pd.read_csv(path + "dRC_Replication/data/ipums_variables_SOC_minor_w_exp.csv", delimiter = ",")
    occ_ids = pd.read_csv("/Users/ebbamark/OneDrive - Nexus365/GenerateOccMobNets/data/soc_2010_minor_codes_employment_asec.csv", delimiter=",")[['SOC']]
    occ_ids = occ_ids.rename(columns={"SOC": "Code"})
    occ_ids['id'] = occ_ids.index.values
    occ_ids['label'] = occ_ids['Code']
    assert(all(occ_ids['id'].values == range(0,len(occ_ids))))
    assert(all(occ_ids.index.values == occ_ids['id'].values))

    employment = np.round(ipums_input[['emp']]/10000) + 1
    unemployment = round(employment * (0.05 / 0.95))
    vac_rate_base = pd.read_csv(path + "dRC_Replication/data/vacancy_rateDec2000.csv").iloc[:, 2].mean() / 100
    vacancies = round(employment * vac_rate_base / (1 - vac_rate_base))
    demand_target = employment + vacancies

    print("Data loaded successfully!")

    # %%
    # Combine into analysis dataframe
    analysis_df = pd.DataFrame({
        'occ_id': range(len(A)),
        'label': occ_ids['label'].values,
        'code': occ_ids['Code'].values,
        'employment': employment.values.flatten(),
        'unemployment': unemployment.values.flatten(),
        'vacancies': vacancies.values.flatten(),
        'target_demand': demand_target.values.flatten(),
        'wage': ipums_input['median_weekly_earnings'].values * 52
    })

    print(f"Created analysis dataframe with {len(analysis_df)} occupations")

# %%
# Calculate network statistics
print("Calculating network statistics...")
A_matrix = A.values
n_neighbors = (A_matrix > 0).sum(axis=1)
neighbor_avg_size = []
neighbor_small_count = []

SMALL_THRESHOLD = 20

for i in range(len(A)):
    neighbors = np.where(A_matrix[i] > 0)[0]
    if len(neighbors) > 0:
        neighbor_sizes = analysis_df.loc[neighbors, 'target_demand'].values
        neighbor_avg_size.append(np.mean(neighbor_sizes))
        neighbor_small_count.append(np.sum(neighbor_sizes < SMALL_THRESHOLD))
    else:
        neighbor_avg_size.append(0)
        neighbor_small_count.append(0)

analysis_df['n_neighbors'] = n_neighbors
analysis_df['neighbor_avg_size'] = neighbor_avg_size
analysis_df['neighbor_small_count'] = neighbor_small_count
analysis_df['is_small'] = analysis_df['target_demand'] < SMALL_THRESHOLD

# Identify bottleneck occupations
analysis_df['bottleneck_risk'] = (
    (analysis_df['is_small']) | 
    (analysis_df['neighbor_avg_size'] < SMALL_THRESHOLD) |
    (analysis_df['neighbor_small_count'] > analysis_df['n_neighbors'] * 0.5)
)

print("Network statistics calculated!")

# %%
# Display summary statistics
print("="*80)
print("NETWORK BOTTLENECK ANALYSIS")
print("="*80)
print(f"\nTotal occupations: {len(analysis_df)}")
print(f"Small occupations (target_demand < {SMALL_THRESHOLD}): {analysis_df['is_small'].sum()} ({analysis_df['is_small'].mean():.1%})")
print(f"Occupations at bottleneck risk: {analysis_df['bottleneck_risk'].sum()} ({analysis_df['bottleneck_risk'].mean():.1%})")
print(f"\nTarget demand distribution:")
print(f"  Min: {analysis_df['target_demand'].min():.0f}")
print(f"  25th percentile: {analysis_df['target_demand'].quantile(0.25):.0f}")
print(f"  Median: {analysis_df['target_demand'].median():.0f}")
print(f"  75th percentile: {analysis_df['target_demand'].quantile(0.75):.0f}")
print(f"  Max: {analysis_df['target_demand'].max():.0f}")

# %%
# Show smallest occupations
print("\n" + "="*80)
print("SMALLEST OCCUPATIONS (Top 20)")
print("="*80)
smallest = analysis_df.nsmallest(20, 'target_demand')[['occ_id', 'label', 'target_demand', 'n_neighbors', 'neighbor_avg_size']]
display(smallest)

# %%
# Show highest bottleneck risk occupations
print("\n" + "="*80)
print("HIGHEST BOTTLENECK RISK OCCUPATIONS")
print("="*80)
bottleneck_occs = analysis_df[analysis_df['bottleneck_risk']].sort_values('target_demand')
print(f"\nOccupations with small neighbors (neighbor_avg_size < {SMALL_THRESHOLD}):")
small_neighbor_occs = bottleneck_occs[bottleneck_occs['neighbor_avg_size'] < SMALL_THRESHOLD].head(15)
display(small_neighbor_occs[['occ_id', 'label', 'target_demand', 'n_neighbors', 'neighbor_small_count', 'neighbor_avg_size']])

# %% [markdown]
# ## Visualization 1: Size Distribution

# %%
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Histogram of target_demand
ax = axes[0, 0]
ax.hist(analysis_df['target_demand'], bins=50, edgecolor='black', alpha=0.7)
ax.axvline(SMALL_THRESHOLD, color='red', linestyle='--', linewidth=2, label=f'Small threshold ({SMALL_THRESHOLD})')
ax.set_xlabel('Target Demand', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Distribution of Occupation Sizes', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)

# Log scale histogram
ax = axes[0, 1]
ax.hist(np.log10(analysis_df['target_demand']), bins=50, edgecolor='black', alpha=0.7, color='orange')
ax.axvline(np.log10(SMALL_THRESHOLD), color='red', linestyle='--', linewidth=2, label=f'Small threshold')
ax.set_xlabel('Log10(Target Demand)', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Distribution of Occupation Sizes (Log Scale)', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)

# Scatter: Own size vs Neighbor average size
ax = axes[1, 0]
colors = ['red' if x else 'blue' for x in analysis_df['bottleneck_risk']]
ax.scatter(analysis_df['target_demand'], analysis_df['neighbor_avg_size'], 
           c=colors, alpha=0.6, s=30)
ax.axhline(SMALL_THRESHOLD, color='red', linestyle='--', linewidth=1, alpha=0.5)
ax.axvline(SMALL_THRESHOLD, color='red', linestyle='--', linewidth=1, alpha=0.5)
ax.set_xlabel('Own Target Demand', fontsize=12)
ax.set_ylabel('Average Neighbor Target Demand', fontsize=12)
ax.set_title('Own Size vs Neighbor Size (Red = Bottleneck Risk)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(alpha=0.3)

# Number of neighbors distribution
ax = axes[1, 1]
ax.hist(analysis_df['n_neighbors'], bins=30, edgecolor='black', alpha=0.7, color='green')
ax.set_xlabel('Number of Neighbors', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Distribution of Network Connectivity', fontsize=14, fontweight='bold')
ax.grid(alpha=0.3)

plt.tight_layout()
# plt.savefig(path + 'output/network_bottleneck_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

# %% [markdown]
# ## Visualization 2: Network Graph with Size-Based Coloring

# %%
print("Creating network graph visualization...")

# Create NetworkX graph (sample if too large)
G = nx.DiGraph()
threshold_connections = 0.01  # Only show edges with weight > 1%

for i in range(len(A)):
    G.add_node(i, size=analysis_df.loc[i, 'target_demand'], 
               label=analysis_df.loc[i, 'label'],
               bottleneck=analysis_df.loc[i, 'bottleneck_risk'])

for i in range(len(A)):
    for j in range(len(A)):
        if A_matrix[i, j] > threshold_connections:
            G.add_edge(i, j, weight=A_matrix[i, j])

# Layout
print("Computing layout (this may take a moment)...")
pos = nx.spring_layout(G, k=0.5, iterations=50, seed=42)

fig, ax = plt.subplots(figsize=(20, 20))

# Node colors based on size
node_colors = []
for node in G.nodes():
    if analysis_df.loc[node, 'target_demand'] < 5:
        node_colors.append('darkred')
    elif analysis_df.loc[node, 'target_demand'] < 10:
        node_colors.append('red')
    elif analysis_df.loc[node, 'target_demand'] < 20:
        node_colors.append('orange')
    elif analysis_df.loc[node, 'target_demand'] < 50:
        node_colors.append('yellow')
    else:
        node_colors.append('lightgreen')

# Node sizes proportional to log(target_demand)
node_sizes = [np.log10(analysis_df.loc[node, 'target_demand'] + 1) * 50 for node in G.nodes()]

# Draw edges
nx.draw_networkx_edges(G, pos, alpha=0.1, width=0.5, arrows=True, 
                       arrowsize=5, ax=ax, edge_color='gray')

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes, 
                       alpha=0.8, ax=ax)

# Add legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='darkred', markersize=10, label='< 5'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='5-10'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=10, label='10-20'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=10, label='20-50'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='lightgreen', markersize=10, label='> 50')
]
ax.legend(handles=legend_elements, title='Target Demand', loc='upper right', fontsize=12)

ax.set_title('Occupational Mobility Network\n(Node color = size, Edge = transition probability > 1%)', 
             fontsize=16, fontweight='bold')
ax.axis('off')

# plt.savefig(path + 'output/network_graph_by_size.png', dpi=300, bbox_inches='tight')
plt.show()

# %% [markdown]
# ## Visualization 3: Adjacency Matrix Heatmaps

# %%
print("Creating adjacency matrix heatmaps...")

fig, axes = plt.subplots(1, 2, figsize=(24, 10))

# Sort by target_demand for first heatmap
sort_idx = analysis_df.sort_values('target_demand').index.values
A_sorted = A_matrix[sort_idx][:, sort_idx]

ax = axes[0]  # First plot
im = ax.imshow(A_sorted, cmap='YlOrRd', aspect='auto', vmin=0, vmax=0.1)
ax.set_title('Adjacency Matrix (Sorted by Size)\nSmall  Large', fontsize=14, fontweight='bold')
ax.set_xlabel('To Occupation', fontsize=12)
ax.set_ylabel('From Occupation', fontsize=12)
plt.colorbar(im, ax=ax, label='Transition Probability')

# Add lines to mark small occupation regions
small_count = (analysis_df['target_demand'] < SMALL_THRESHOLD).sum()
ax.axhline(small_count, color='red', linestyle='--', linewidth=2, alpha=0.7)
ax.axvline(small_count, color='red', linestyle='--', linewidth=2, alpha=0.7)
ax.text(5, small_count - 5, f'Small occs (n={small_count})', color='red', fontsize=10, fontweight='bold')

# Hierarchical clustering of network structure
ax = axes[1]  # CHANGED: Second plot (was axes[0] before)

# Alternative: Use correlation-based distance
from scipy.spatial.distance import pdist, squareform as sq

# Compute pairwise correlation between occupation transition patterns
correlation_matrix = np.corrcoef(A_matrix)

# Convert correlation to distance (1 - correlation)
dist_matrix = 1 - correlation_matrix
np.fill_diagonal(dist_matrix, 0)

# Ensure non-negative and convert to condensed form
dist_matrix = np.clip(dist_matrix, 0, None)
condensed_dist = pdist(A_matrix, metric='correlation')

# Hierarchical clustering
linkage_matrix = linkage(condensed_dist, method='ward')
cluster_order = dendrogram(linkage_matrix, no_plot=True)['leaves']

A_clustered = A_matrix[cluster_order][:, cluster_order]

im = ax.imshow(A_clustered, cmap='YlOrRd', aspect='auto', vmin=0, vmax=0.1)
ax.set_title('Adjacency Matrix (Hierarchical Clustering)\nShows Network Communities', fontsize=14, fontweight='bold')
ax.set_xlabel('To Occupation', fontsize=12)
ax.set_ylabel('From Occupation', fontsize=12)
plt.colorbar(im, ax=ax, label='Transition Probability')

plt.tight_layout()
# plt.savefig(path + 'output/network_adjacency_heatmaps.png', dpi=300, bbox_inches='tight')
plt.show()

# Identify small occupations and their clustering
small_occs = analysis_df[analysis_df['is_small']].copy()

# For each small occupation, check if it's in a cluster with other small occupations
small_occs['n_small_neighbors'] = small_occs['neighbor_small_count']
small_occs['pct_small_neighbors'] = small_occs['neighbor_small_count'] / small_occs['n_neighbors']

# Identify the most isolated small occupations
most_isolated = small_occs.nlargest(20, 'pct_small_neighbors')[
    ['occ_id', 'label', 'target_demand', 'n_neighbors', 'n_small_neighbors', 'pct_small_neighbors']
]

print("\n" + "="*80)
print("MOST ISOLATED SMALL OCCUPATIONS")
print("(High % of neighbors are also small)")
print("="*80)
display(most_isolated)

# Count how many small occupations are in "small clusters"
high_risk_isolated = (small_occs['pct_small_neighbors'] > 0.5).sum()
print(f"\nSmall occupations with >50% small neighbors: {high_risk_isolated} / {len(small_occs)} ({high_risk_isolated/len(small_occs):.1%})")

# Extract just the small occupations and their connections
small_occ_ids = analysis_df[analysis_df['is_small']].index.values
A_small = A_matrix[small_occ_ids][:, small_occ_ids]

fig, ax = plt.subplots(figsize=(12, 10))
im = ax.imshow(A_small, cmap='YlOrRd', aspect='auto', vmin=0, vmax=0.1)
ax.set_title(f'Connections ONLY Among Small Occupations (n={len(small_occ_ids)})\nShows Isolated Small Clusters', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('To Small Occupation', fontsize=12)
ax.set_ylabel('From Small Occupation', fontsize=12)
plt.colorbar(im, ax=ax, label='Transition Probability')
plt.tight_layout()
plt.show()

# Calculate how connected small occupations are to each other
small_internal_connections = A_small.sum() / (len(small_occ_ids) ** 2)
print(f"Average transition probability between small occupations: {small_internal_connections:.4f}")
print(f"This is {'LOW' if small_internal_connections < 0.01 else 'MODERATE' if small_internal_connections < 0.03 else 'HIGH'}")
# %% [markdown]
# ## Visualization 4: Bottleneck Risk Analysis

# %%
print("Creating bottleneck risk map...")

fig, axes = plt.subplots(2, 2, figsize=(18, 14))

# Map 1: Own size vs neighbors with small count
ax = axes[0, 0]
scatter = ax.scatter(analysis_df['target_demand'], analysis_df['n_neighbors'],
                    c=analysis_df['neighbor_small_count'], s=100, alpha=0.6, cmap='Reds')
ax.axvline(SMALL_THRESHOLD, color='blue', linestyle='--', linewidth=2, alpha=0.5, label='Small threshold')
ax.set_xlabel('Own Target Demand', fontsize=12)
ax.set_ylabel('Number of Neighbors', fontsize=12)
ax.set_title('Occupation Size vs Connectivity\n(Color = Number of Small Neighbors)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.legend()
ax.grid(alpha=0.3)
plt.colorbar(scatter, ax=ax, label='Small Neighbors Count')

# Map 2: Bottleneck risk distribution
ax = axes[0, 1]
bottleneck_colors = ['red' if x else 'lightblue' for x in analysis_df['bottleneck_risk']]
ax.scatter(analysis_df['target_demand'], analysis_df['neighbor_avg_size'],
          c=bottleneck_colors, s=50, alpha=0.6)
ax.axhline(SMALL_THRESHOLD, color='red', linestyle='--', linewidth=1.5, alpha=0.5)
ax.axvline(SMALL_THRESHOLD, color='red', linestyle='--', linewidth=1.5, alpha=0.5)
ax.fill_between([0, SMALL_THRESHOLD], 0, SMALL_THRESHOLD, alpha=0.2, color='red', label='High Risk Zone')
ax.set_xlabel('Own Target Demand', fontsize=12)
ax.set_ylabel('Average Neighbor Size', fontsize=12)
ax.set_title('Bottleneck Risk Map\n(Red = At Risk)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()
ax.grid(alpha=0.3)

# Map 3: Network density by size category
ax = axes[1, 0]
size_categories = pd.cut(analysis_df['target_demand'], bins=[0, 5, 10, 20, 50, np.inf], 
                         labels=['<5', '5-10', '10-20', '20-50', '>50'])
analysis_df['size_cat'] = size_categories

connectivity_by_size = analysis_df.groupby('size_cat')['n_neighbors'].agg(['mean', 'std', 'count'])
x_pos = range(len(connectivity_by_size))
ax.bar(x_pos, connectivity_by_size['mean'], yerr=connectivity_by_size['std'], 
       color='steelblue', alpha=0.7, capsize=5)
ax.set_xticks(x_pos)
ax.set_xticklabels(connectivity_by_size.index)
ax.set_xlabel('Size Category', fontsize=12)
ax.set_ylabel('Average Number of Neighbors', fontsize=12)
ax.set_title('Network Connectivity by Occupation Size', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Add count labels
for i, (idx, row) in enumerate(connectivity_by_size.iterrows()):
    ax.text(i, row['mean'] + row['std'] + 1, f"n={row['count']:.0f}", 
            ha='center', fontsize=10)

# Map 4: Neighbor quality distribution
ax = axes[1, 1]
small_occs = analysis_df[analysis_df['is_small']]
large_occs = analysis_df[~analysis_df['is_small']]

ax.hist(small_occs['neighbor_avg_size'], bins=30, alpha=0.6, label=f'Small occs (n={len(small_occs)})', 
        color='red', edgecolor='black')
ax.hist(large_occs['neighbor_avg_size'], bins=30, alpha=0.6, label=f'Large occs (n={len(large_occs)})', 
        color='blue', edgecolor='black')
ax.axvline(SMALL_THRESHOLD, color='black', linestyle='--', linewidth=2, label='Risk threshold')
ax.set_xlabel('Average Neighbor Size', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Distribution of Neighbor Quality\nSmall vs Large Occupations', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
# plt.savefig(path + 'output/bottleneck_risk_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# %% [markdown]
# ## Visualization 5: Expected Flow Analysis

# %%
print("Creating expected flow analysis...")

vr_t = 0.03  # 3% vacancy rate
sep_rate = 0.02  # 2% separation rate

analysis_df['expected_vacs'] = analysis_df['target_demand'] * vr_t
analysis_df['expected_seps'] = analysis_df['employment'] * sep_rate
analysis_df['prob_zero_vacs'] = (1 - vr_t) ** analysis_df['target_demand']
analysis_df['prob_zero_seps'] = (1 - sep_rate) ** analysis_df['employment']

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Expected vacancies vs size
ax = axes[0, 0]
ax.scatter(analysis_df['target_demand'], analysis_df['expected_vacs'], 
          c=analysis_df['prob_zero_vacs'], s=50, alpha=0.6, cmap='RdYlGn_r')
ax.axhline(1, color='red', linestyle='--', linewidth=2, label='1 vacancy threshold')
ax.set_xlabel('Target Demand', fontsize=12)
ax.set_ylabel('Expected Vacancies (vr=3%)', fontsize=12)
ax.set_title('Expected Vacancy Creation by Size\n(Color = P(0 vacancies))', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.legend()
ax.grid(alpha=0.3)
plt.colorbar(ax.collections[0], ax=ax, label='P(Zero Vacancies)')

# Probability of zero vacancies
ax = axes[0, 1]
ax.scatter(analysis_df['target_demand'], analysis_df['prob_zero_vacs'] * 100,
          c=analysis_df['bottleneck_risk'].map({True: 'red', False: 'blue'}),
          s=50, alpha=0.6)
ax.set_xlabel('Target Demand', fontsize=12)
ax.set_ylabel('Probability of Zero Vacancies (%)', fontsize=12)
ax.set_title('Risk of No Vacancy Creation\n(Red = Bottleneck Risk)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.grid(alpha=0.3)

# Expected separations vs size
ax = axes[1, 0]
ax.scatter(analysis_df['employment'], analysis_df['expected_seps'],
          c=analysis_df['prob_zero_seps'], s=50, alpha=0.6, cmap='RdYlGn_r')
ax.axhline(1, color='red', linestyle='--', linewidth=2, label='1 separation threshold')
ax.set_xlabel('Employment', fontsize=12)
ax.set_ylabel('Expected Separations (rate=2%)', fontsize=12)
ax.set_title('Expected Separations by Size\n(Color = P(0 separations))', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.legend()
ax.grid(alpha=0.3)
plt.colorbar(ax.collections[0], ax=ax, label='P(Zero Separations)')

# Imbalance: expected vacs vs seps
ax = axes[1, 1]
analysis_df['flow_imbalance'] = analysis_df['expected_vacs'] - analysis_df['expected_seps']
colors = ['red' if x < 0 else 'green' if x > 0.5 else 'yellow' for x in analysis_df['flow_imbalance']]
ax.scatter(analysis_df['target_demand'], analysis_df['flow_imbalance'],
          c=colors, s=50, alpha=0.6)
ax.axhline(0, color='black', linestyle='-', linewidth=1)
ax.set_xlabel('Target Demand', fontsize=12)
ax.set_ylabel('Expected Vacs - Expected Seps', fontsize=12)
ax.set_title('Flow Imbalance by Size\n(Green=Surplus Vacs, Red=Surplus Seps)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.grid(alpha=0.3)

plt.tight_layout()
# plt.savefig(path + 'output/expected_flow_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# %% [markdown]
# ## Summary and Export Options

# %%
print("\n" + "="*80)
print("ANALYSIS COMPLETE!")
print("="*80)
print("\nKey Findings:")
print(f"- {analysis_df['is_small'].sum()} occupations have target_demand < {SMALL_THRESHOLD}")
print(f"- {analysis_df['bottleneck_risk'].sum()} occupations are at bottleneck risk")
print(f"- {(analysis_df['prob_zero_vacs'] > 0.5).sum()} occupations have >50% chance of creating 0 vacancies")
print(f"- Smallest occupation size: {analysis_df['target_demand'].min():.0f}")
print(f"- Median occupation size: {analysis_df['target_demand'].median():.0f}")

# %%
# Optional: Export bottleneck occupations for further analysis
# Uncomment to save to CSV
"""
bottleneck_export = analysis_df[analysis_df['bottleneck_risk']].sort_values('target_demand')
bottleneck_export.to_csv(path + 'output/bottleneck_occupations.csv', index=False)
print("Bottleneck occupations exported to CSV")
"""

# %%
```

```python
"""
Simple function to add to your existing network_diagnostics notebook.
Just add this function and call it with your network data and observation data.
"""

def plot_isolation_vs_ltuer(network_data, observation, agg_level='SOC2010_adj', 
                            save=False, path=None):
    """
    Create a single diagnostic plot showing the relationship between 
    network isolation metrics and long-term unemployment rates.
    
    Parameters:
    -----------
    network_data : DataFrame
        Network analysis data with columns: occ_id, label, target_demand, 
        n_neighbors, neighbor_avg_size
    observation : DataFrame
        Observation data with LTUER values
    agg_level : str
        SOC aggregation level ('SOC2010_adj', 'SOC_broad_adj', etc.)
    save : bool
        Whether to save the figure
    path : str
        Path to save figure (if save=True)
    """
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy import stats
    
    # Map aggregation levels to LTUER columns
    ltuer_mapping = {
        'code': 'ltuer_occ',
        'SOC2010_adj': 'ltuer_soc2010',
        'SOC_broad_adj': 'ltuer_soc_broad',
        'SOC_minor_adj': 'ltuer_soc_minor',
        'SOC_major_adj': 'ltuer_soc_major'
    }
    
    ltuer_col = ltuer_mapping.get(agg_level, 'ltuer_occ')
    
    # Get unique LTUER values for each occupation code
    ltuer_data = observation[[agg_level, ltuer_col]].drop_duplicates(subset=[agg_level])
    
    # Merge with network data
    # Try to merge on the label column (adjust if needed based on your data structure)
    merged = network_data.merge(
        ltuer_data, 
        left_on='code',  # or use 'occ_id' or another column as appropriate
        right_on=agg_level,
        how='inner'
    )
    
    if len(merged) == 0:
        print(f"Warning: No matches found when merging. Check column names.")
        print(f"Network columns: {network_data.columns.tolist()}")
        print(f"Observation columns: {observation.columns.tolist()}")
        return None
    
    # Calculate a simple isolation score
    # Higher score = more isolated (small size, few neighbors, small neighbors)
    merged['isolation_score'] = (
        1 / (merged['target_demand'] + 1) +
        1 / (merged['n_neighbors'] + 1) + 
        1 / (merged['neighbor_avg_size'] + 1)
    ) / 3
    
    # Create figure with 4 subplots
    fig = plt.figure(figsize=(16, 12))
    gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)
    
    # Define metrics to plot
    metrics = [
        ('target_demand', 'Occupation Size\n(Target Demand)', True),  # Inverse relationship
        ('n_neighbors', 'Number of\nNeighbors', True),
        ('neighbor_avg_size', 'Average\nNeighbor Size', True),
        ('isolation_score', 'Composite\nIsolation Score', False)
    ]
    
    # Plot each metric
    for idx, (metric, label, log_scale) in enumerate(metrics):
        row = idx // 2
        col = idx % 2
        ax = fig.add_subplot(gs[row, col])
        
        # Scatter plot
        ax.scatter(merged[metric], merged[ltuer_col], 
                  alpha=0.6, s=80, edgecolors='black', linewidth=0.5)
        
        # Trend line
        if len(merged) > 2:
            z = np.polyfit(merged[metric], merged[ltuer_col], 1)
            p = np.poly1d(z)
            x_line = np.linspace(merged[metric].min(), merged[metric].max(), 100)
            ax.plot(x_line, p(x_line), "r--", alpha=0.8, linewidth=2.5, 
                   label=f'Trend: y={z[0]:.4f}x+{z[1]:.3f}')
        
        # Calculate correlation
        corr, p_value = stats.pearsonr(merged[metric], merged[ltuer_col])
        
        # Determine significance
        if p_value < 0.001:
            sig = "***"
        elif p_value < 0.01:
            sig = "**"
        elif p_value < 0.05:
            sig = "*"
        else:
            sig = "ns"
        
        # Labels and title
        ax.set_xlabel(label, fontsize=12, fontweight='bold')
        ax.set_ylabel(f'Long-Term Unemployment Rate\n({agg_level})', fontsize=11)
        ax.set_title(f'Correlation: r = {corr:.3f} ({sig})\np-value: {p_value:.4f}', 
                    fontsize=11)
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=9)
        
        if log_scale and merged[metric].min() > 0:
            ax.set_xscale('log')
    
    # Add risk matrix in the bottom right quadrant
    ax_matrix = fig.add_subplot(gs[:, 2])
    
    # Create risk quadrant plot
    iso_median = merged['isolation_score'].median()
    ltuer_median = merged[ltuer_col].median()
    
    # Color code by risk level
    colors = []
    for _, row in merged.iterrows():
        if row['isolation_score'] > iso_median and row[ltuer_col] > ltuer_median:
            colors.append('#d62728')  # Red - high risk
        elif row['isolation_score'] > iso_median:
            colors.append('#ff7f0e')  # Orange - isolated only
        elif row[ltuer_col] > ltuer_median:
            colors.append('#bcbd22')  # Yellow-green - high LTUER only
        else:
            colors.append('#2ca02c')  # Green - low risk
    
    scatter = ax_matrix.scatter(merged['isolation_score'], merged[ltuer_col],
                               c=colors, s=100, alpha=0.7, edgecolors='black', linewidth=0.5)
    
    # Add quadrant lines
    ax_matrix.axvline(iso_median, color='gray', linestyle='--', alpha=0.6, linewidth=2)
    ax_matrix.axhline(ltuer_median, color='gray', linestyle='--', alpha=0.6, linewidth=2)
    
    # Add quadrant labels
    ax_matrix.text(0.05, 0.95, 'High LTUER\nLow Isolation', 
                  transform=ax_matrix.transAxes, fontsize=10, 
                  verticalalignment='top', bbox=dict(boxstyle='round', 
                  facecolor='#bcbd22', alpha=0.3))
    ax_matrix.text(0.95, 0.95, 'HIGH RISK\nIsolated +\nHigh LTUER', 
                  transform=ax_matrix.transAxes, fontsize=10, 
                  verticalalignment='top', horizontalalignment='right',
                  bbox=dict(boxstyle='round', facecolor='#d62728', alpha=0.3),
                  fontweight='bold')
    ax_matrix.text(0.05, 0.05, 'LOW RISK', 
                  transform=ax_matrix.transAxes, fontsize=10,
                  verticalalignment='bottom',
                  bbox=dict(boxstyle='round', facecolor='#2ca02c', alpha=0.3),
                  fontweight='bold')
    ax_matrix.text(0.95, 0.05, 'Isolated\nLow LTUER', 
                  transform=ax_matrix.transAxes, fontsize=10,
                  verticalalignment='bottom', horizontalalignment='right',
                  bbox=dict(boxstyle='round', facecolor='#ff7f0e', alpha=0.3))
    
    ax_matrix.set_xlabel('Composite Isolation Score', fontsize=12, fontweight='bold')
    ax_matrix.set_ylabel(f'Long-Term Unemployment Rate\n({agg_level})', fontsize=11)
    ax_matrix.set_title('Risk Matrix', fontsize=12, fontweight='bold')
    ax_matrix.grid(True, alpha=0.3)
    
    # Count occupations in each quadrant
    high_risk = len(merged[(merged['isolation_score'] > iso_median) & 
                           (merged[ltuer_col] > ltuer_median)])
    low_risk = len(merged[(merged['isolation_score'] <= iso_median) & 
                          (merged[ltuer_col] <= ltuer_median)])
    
    # Add summary text
    summary_text = f"High Risk: {high_risk} ({high_risk/len(merged)*100:.1f}%)\n"
    summary_text += f"Low Risk: {low_risk} ({low_risk/len(merged)*100:.1f}%)\n"
    summary_text += f"n = {len(merged)} occupations"
    
    ax_matrix.text(0.5, -0.15, summary_text, transform=ax_matrix.transAxes,
                  fontsize=10, ha='center', 
                  bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    # Overall title
    fig.suptitle(f'Network Isolation vs Long-Term Unemployment Rate ({agg_level})',
                fontsize=16, fontweight='bold', y=0.98)
    
    # Print summary statistics
    print("="*80)
    print(f"NETWORK ISOLATION vs LTUER ANALYSIS ({agg_level})")
    print("="*80)
    #print(f"\nSample size: {len(merged)} occupations")
    print(f"\nIsolation Score: mean={merged['isolation_score'].mean():.3f}, "
          f"median={merged['isolation_score'].median():.3f}")
    print(f"LTUER: mean={merged[ltuer_col].mean():.3f}, "
          f"median={merged[ltuer_col].median():.3f}")
    print(f"\nOverall correlation: r={stats.pearsonr(merged['isolation_score'], merged[ltuer_col])[0]:.3f}")
    print(f"\nHigh-risk occupations (isolated + high LTUER): {high_risk} ({high_risk/len(merged)*100:.1f}%)")
    print("="*80)
    
    #if save and path:
        #fig.savefig(f'{path}isolation_vs_ltuer_{agg_level}.png', dpi=300, bbox_inches='tight')
        #print(f"\nFigure saved to: {path}isolation_vs_ltuer_{agg_level}.png")
    
    plt.show()
    
    return merged


# Example usage - add this cell to your notebook:
occ_ltuer_obs = pd.read_csv(path + "data/occ_uer_ltuer_observed.csv",
    dtype={
        'acs_occ_code':int,
            'SOC_major_adj': str,
            'SOC_minor_adj': str,
            'SOC_broad_adj': str,
            'SOC2010_adj': str,
        }
    ).rename(columns={'acs_occ_code': 'code'})
# After running your network analysis, call:
merged_data = plot_isolation_vs_ltuer(
    network_data=analysis_df,  # Your network DataFrame with isolation metrics
    observation=occ_ltuer_obs,    # Your observation DataFrame with LTUER
    agg_level='code',   # Choose aggregation level
    save=True,                 # Set to True to save figure
    path='./outputs/'          # Path to save (if save=True)
)
print(occ_ltuer_obs)

#print(analysis_df)
```
